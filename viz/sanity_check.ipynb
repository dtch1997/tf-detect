```python
import os
import tensorflow as tf
os.chdir("/home/dtch009/tf-detect")

MODEL_NAME = "ssd_mobilenet_v2_96x96_depth=0.2"

converter = tf.lite.TFLiteConverter.from_saved_model(f'exported_models/{MODEL_NAME}/saved_model', signature_keys=['serving_default'])
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.experimental_new_converter = True
converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS, tf.lite.OpsSet.SELECT_TF_OPS]
quantized_model = converter.convert()
open(f'exported_models/{MODEL_NAME}/model.tflite', "wb").write(quantized_model)
```
